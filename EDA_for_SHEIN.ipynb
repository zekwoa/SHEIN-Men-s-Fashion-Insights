```python
import pandas as pd
import re


df = pd.read_excel('shein_mens_fashion.xlsx')


def extract_property(description, property_name):

    match = re.search(rf'{property_name}:\s*([A-Za-z\s]+)', description)
    if match:
        return match.group(1).strip()  # إرجاع القيمة بعد الكلمة المحددة
    return None

df['Color'] = df['description'].apply(lambda x: extract_property(x, 'Color'))


df['Style'] = df['description'].apply(lambda x: extract_property(x, 'Style'))

df['Pattern Type'] = df['description'].apply(lambda x: extract_property(x, 'Pattern Type'))

df['Details'] = df['description'].apply(lambda x: extract_property(x, 'Details'))

df['Top Type'] = df['description'].apply(lambda x: extract_property(x, 'Top Type'))

df['Bottom Type'] = df['description'].apply(lambda x: extract_property(x, 'Bottom Type'))

df['Neckline'] = df['description'].apply(lambda x: extract_property(x, 'Neckline'))

df['Sleeve Type'] = df['description'].apply(lambda x: extract_property(x, 'Sleeve Type'))

df['Fit Type'] = df['description'].apply(lambda x: extract_property(x, 'Fit Type'))

df['Number of Pieces'] = df['description'].apply(lambda x: extract_property(x, 'Number of Pieces'))

df['Fabric'] = df['description'].apply(lambda x: extract_property(x, 'Fabric'))

df['Care Instructions'] = df['description'].apply(lambda x: extract_property(x, 'Care Instructions'))

print(df[['description', 'Color', 'Style', 'Pattern Type', 'Details','Top Type' , 'Bottom Type' , 'Neckline',
 'Sleeve Type' , 'Fit Type' ,'Number of Pieces' , 'Fabric' , 'Care Instructions' ]])
```
